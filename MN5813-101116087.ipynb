### Project summary

Within the professional football context, injuries not only affect athletic performance but also creates a significant impact in economic terms: salaries paid during days off, limited availability of key players and even may lead to poor performance. Nevertheless, the idea of quantifying that impact is often challenging because the information is widely distributed across multiple sources (clubs and leagues data, injury records, salaries) and there is not always a clean join between datasets.
This project analyzes recent injury records and combines them with salary information in order to estimate the expected economic cost associated with the absence of players due to injuries, by comparing patterns across leagues and clubs


### Aim

To estimate and compare the impact of professional football injuries by merging injury and salary data to describe patterns according to the seasons, league, club, injury type, and provide an economic cost estimate associated with days off.

### Objectives

-To collect, clean and combine datasets regarding injuries, clubs, competitions and salaries, documenting assumptions, inclusion and exclusion criteria according to accepted definitions in the literature. 

-To quantify the frequency of injuries and severity (days off) and explore its distribution by season, league, club and injury type aligned with metrics used in soccer injury surveillance.

-To estimate an economic impact linked to injuries by using a wage and days off approach, comparing aggregated results by league and club and discussing the implications of sports management and investment prevention.

### Review

Injuries in professional football have been investigated for years due to the fact that they are not ‚Äúisolated‚Äù events, but a recurring phenomenon that repeats itself season after season. A reliable indicator of the magnitude of the presented topic is that long-term surveillance programs such as UEFA Elite Club Injury Study have acquired injury data for more than two decades, which reinforces the problem is frequent enough to demand systematic monitoring over time (Ekstrand et al., 2022)

In recent years, the problem has become notably visible due to its annual scale. Howden‚Äôs Men‚Äôs European Football Injury Index reports that during season 2023/2024 (top five European leagues) there was a maximum of 4,123 injuries and an estimated cost of ‚Ç¨732.02 million in wages paid to players while being inactive. That same report sustains that the frequency and costs have been increasing since 2020/2021 and that, since the index‚Äôs first edition, injuries have accumulated to 14,292 injuries (excluding COVID-19) with an approximate cost of ‚Ç¨2.3 billion in wages (Howden Group Holdings, 2024).

When considering the multi-season picture, the matter becomes particularly compelling. In the 2024/2025 index edition (five-season review), Howden indicates 22,596 injuries between years 2020 and 2025 within Europe‚Äôs five top leagues, with an estimated wage cost of ‚Ç¨3.45 billion. This establishes injuries as not only an athletic issue but also a financial risk management one: clubs invest substantially in talent, but a portion of that investment is temporarily ‚Äúlost‚Äù when a player is during a day off (Howden Group Holdings, 2024).

In terms of possible causes, the literature and public reports typically refrain from single explanations, but there is a context that repetitively appears: match load and calendar congestion. Through FIFPRO‚Äôs monitoring system (‚âà1,500 players), it was reported that 54% of the observed players faced ‚Äúexcessive or high‚Äù load demands in 2023/2024 and connects this pattern to risks to health and performance when recovery is deficient (FIFPRO, 2024).

Against this backdrop, the present project matches with what can be transparently measured using public available data: quantifying how many injuries are recorded (counts by season, league and injury type), describing their severity using days of absence (when available in dataset) and approximating the direct economic impact as an estimate based on wages during periods of absence, in line with the ‚Äúwage costs‚Äù approach used in recent sector reports.




### Data Source

This project integrates two types of sources to study injuries and estimate its economic impact:
Injury records (2020-2025): it is extracted from the European Football Injuries (2020-2025) dataset published on Kaggle. This information contains event-level injury records with multiple variables such as season, club, league, injury type and days off (Muzaffarov, 2026).
Salaries by league: salary tables by player are available on FootyStats for the five major European leagues (Premier League, La Liga, Bundesliga, Ligue 1 and Serie A) are used to estimate the direct cost of injuries through salarie.


### Data cleaning

Prior to the analysis, a systematic data cleaning process was carried out to ensure consistency, comparability, and quality of the information:

Data types were reviewed, and key variables (e.g., days off and salaries) were converted to numeric format, handling invalid values ‚Äã‚Äãby setting them to null.

Records with missing essential information (e.g., injuries without a number of days off or players without an identified club) were removed, as these do not allow for a reliable analysis of severity or impact.

Textual variables such as club, league, and player names were standardized through string normalization (conversion to lowercase, removal of spaces and special characters), reducing inconsistencies between databases from different sources.

Duplicates were checked, and only the records necessary for the descriptive analysis were retained.

This process ensures that the data used consistently reflects injury events and minimizes errors arising from inconsistencies in the original record.

### Data preprocessing and integration

Once the data was cleaned, the necessary preprocessing for integration between databases was performed. This involved constructing a join key based on the player's name and league, which allowed linking the injury database with the salary databases.

Subsequently, relevant derived variables for the analysis were generated, such as the estimated daily salary and an approximation of the economic cost of each injury, calculated from the number of days off and the player's salary. These transformations allow for analyzing not only the number of injuries that occur, but also their severity and potential financial impact.

Finally, the data were aggregated by season, league, club, and injury type using grouping operations, which facilitates descriptive analysis and the visualization of general patterns in the dataset.

### Data Cleaning & Management

```python
# =========================
# CELL 4 ‚Äî UNIFY ALL SALARY DATASETS INTO ONE TABLE
# Purpose: Stack (concatenate) all salary DataFrames into a single DataFrame with league + season labels.
# Output: df_salarios (one big salary table)
# =========================

def clean_cols(df: pd.DataFrame) -> pd.DataFrame:  # Basic column cleanup for consistency
    df = df.copy()  # Avoid modifying the original DataFrame
    df = df.loc[:, ~df.columns.astype(str).str.contains(r"^Unnamed", regex=True)]  # Drop "Unnamed" index columns
    df.columns = df.columns.astype(str).str.strip()  # Trim spaces in column names
    df.columns = df.columns.str.replace(r"\s+", " ", regex=True)  # Collapse multiple spaces in column names
    return df  # Return cleaned DataFrame

# List of (df_key_in_dfs, league_label, season_label)
salary_sources = [
    ("bundesliga_22_23", "Bundesliga", "2022-23"),
    ("bundesliga_23_24", "Bundesliga", "2023-24"),
    ("bundesliga_24_25", "Bundesliga", "2024-25"),

    ("premier_22_23", "Premier League", "2022-23"),
    ("premier_23_24", "Premier League", "2023-24"),
    ("premier_24_25", "Premier League", "2024-25"),

    ("laliga_22_23", "La Liga", "2022-23"),
    ("laliga_23_24", "La Liga", "2023-24"),
    ("laliga_24_25", "La Liga", "2024-25"),

    ("seriea_22_23", "Serie A", "2022-23"),
    ("seriea_23_24", "Serie A", "2023-24"),
    ("seriea_24_25", "Serie A", "2024-25"),

    ("ligue1_22_23", "Ligue 1", "2022-23"),
    ("ligue1_23_24", "Ligue 1", "2023-24"),
    ("ligue1_24_25", "Ligue 1", "2024-25"),
]

salary_frames = []  # We will store cleaned + labeled salary tables here

for key, league, season in salary_sources:  # Loop through each salary dataset we want to stack
    df = dfs.get(key)  # Get the DataFrame from the dfs dictionary
    if df is None:
        print(f"‚ö†Ô∏è Skipping {key} (not loaded)")  # Skip if it failed earlier
        continue

    tmp = clean_cols(df)  # Clean column names / remove unnamed cols
    tmp["league"] = league  # Add league label (needed for analysis later)
    tmp["season_source"] = season  # Add season label (from file source)
    salary_frames.append(tmp)  # Add to list for concatenation

# Concatenate into one DataFrame (stack rows)
df_salarios = pd.concat(salary_frames, ignore_index=True, sort=False)  # One big salaries table

print(f"‚úÖ df_salarios created -> rows: {df_salarios.shape[0]:,} | cols: {df_salarios.shape[1]:,}")  # Confirm size
display(df_salarios.head(5))  # Quick preview
```

Salary data from different leagues and seasons were consolidated into a single table (df_salarios) through vertical concatenation. Prior to joining, basic column cleaning was employed in order to eliminate export artefacts (for instance, rows Unnamed) and standardize variable names. Then, league and season_source explicit tags were added to maintain traceability of the origin of each record and allow for comparable analysis by league/season, as well as to enable later integration with the injury database.

```python
# STEP ‚Äî Convert season to "YY/YY" format (e.g., 2022/2023 -> 22/23)

def season_to_yy(x):  # Convert many season formats into "YY/YY"
    if pd.isna(x):
        return np.nan
    s = str(x).strip()

    m = re.match(r"^(\d{4})[\/\-](\d{4})$", s)  # 2022/2023 or 2022-2023
    if m:
        return f"{m.group(1)[-2:]}/{m.group(2)[-2:]}"

    m = re.match(r"^(\d{4})[\/\-](\d{2})$", s)  # 2022-23 or 2022/23
    if m:
        return f"{m.group(1)[-2:]}/{m.group(2)}"

    m = re.match(r"^(\d{2})[\/\-](\d{2})$", s)  # already 22/23
    if m:
        return f"{m.group(1)}/{m.group(2)}"

    return s  # fallback

print("Unique seasons BEFORE:", sorted(df_salarios_clean["season"].astype(str).unique())[:10])  # quick sample
df_salarios_clean["season"] = df_salarios_clean["season"].apply(season_to_yy)  # apply conversion
print("Unique seasons AFTER :", sorted(df_salarios_clean["season"].astype(str).unique())[:10])  # quick sample

display(df_salarios_clean.head(10))  # preview
```

```python
# =========================
# STEP ‚Äî Normalize names (players, clubs, leagues) in BOTH datasets
# Purpose: Remove accents, lowercase, remove punctuation, collapse spaces -> reliable joins.
# Outputs:
#   - df_lesiones_fisicas: player_std, club_std, league_std
#   - df_salarios_clean  : player_std, club_std, league_std
# =========================

def normalize_text(x):  # Clean and standardize text for matching/joins
    if pd.isna(x):  # Keep NaN as NaN
        return np.nan

    s = str(x).strip().lower()  # Trim spaces + lowercase

    # Remove accents/diacritics (e.g., "Aar√≥n" -> "aaron")
    s = unicodedata.normalize("NFKD", s)
    s = "".join(c for c in s if not unicodedata.combining(c))

    # Replace common separators
    s = s.replace("&", " and ")  # Normalize "&" into "and"
    s = re.sub(r"[‚Äô'`¬¥]", " ", s)  # Remove apostrophes
    s = re.sub(r"[-_/]", " ", s)  # Replace separators with spaces

    # Remove everything except letters/numbers/spaces
    s = re.sub(r"[^a-z0-9\s]", " ", s)

    # Collapse multiple spaces
    s = re.sub(r"\s+", " ", s).strip()

    return s

# ---- Apply to injuries table ----
df_lesiones_fisicas["player_std"] = df_lesiones_fisicas["player_name"].apply(normalize_text)  # Standard player name
df_lesiones_fisicas["club_std"]   = df_lesiones_fisicas["club"].apply(normalize_text)  # Standard club name
df_lesiones_fisicas["league_std"] = df_lesiones_fisicas["league"].apply(normalize_text)  # Standard league name

# ---- Apply to salaries table ----
df_salarios_clean["player_std"] = df_salarios_clean["player"].apply(normalize_text)  # Standard player name
df_salarios_clean["club_std"]   = df_salarios_clean["club"].apply(normalize_text)  # Standard club name
df_salarios_clean["league_std"] = df_salarios_clean["league"].apply(normalize_text)  # Standard league name

# ---- Optional: unify league naming (LaLiga vs La Liga, etc.) ----
league_map = {
    "la liga": "laliga",
    "laliga": "laliga",
    "premier league": "premier league",
    "bundesliga": "bundesliga",
    "serie a": "serie a",
    "ligue 1": "ligue 1",
}

df_lesiones_fisicas["league_std"] = df_lesiones_fisicas["league_std"].replace(league_map)  # Canonical league labels
df_salarios_clean["league_std"]   = df_salarios_clean["league_std"].replace(league_map)  # Canonical league labels

# ---- Quick sanity checks ----
print("‚úÖ Injuries std columns added:", ["player_std", "club_std", "league_std"])
print("‚úÖ Salaries std columns added:", ["player_std", "club_std", "league_std"])

print("\nUnique leagues (injuries) BEFORE:", sorted(df_lesiones_fisicas["league"].dropna().unique()))
print("Unique leagues (injuries) AFTER :", sorted(df_lesiones_fisicas["league_std"].dropna().unique()))

print("\nUnique leagues (salaries) BEFORE:", sorted(df_salarios_clean["league"].dropna().unique()))
print("Unique leagues (salaries) AFTER :", sorted(df_salarios_clean["league_std"].dropna().unique()))

display(df_lesiones_fisicas[["player_name","player_std","club","club_std","league","league_std"]].head(8))
display(df_salarios_clean[["player","player_std","club","club_std","league","league_std"]].head(8))
```

Since the injury and salary databases come from different sources, a systematic normalization of key text fields (player, club, and league) was performed in both datasets to reduce integration errors due to spelling differences. Normalization included conversion to lowercase, removal of accents/diacritics, cleaning of punctuation and separators, and collapse of multiple spaces. This procedure improves consistency in joins and reduces the risk of mismatches due to minor variations in the text.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid")

# =========================
# DASHBOARD ‚Äî All charts in one figure (small multiples)
# Requires: df_plot already cleaned + numeric
# =========================

# ---- Aggregations (reused across plots) ----
league_cost = (df_plot.groupby("league", as_index=False)["injury_cost_eur"]
               .sum()
               .sort_values("injury_cost_eur", ascending=False))

top_clubs = (df_plot.groupby("club", as_index=False)["injury_cost_eur"]
             .sum()
             .sort_values("injury_cost_eur", ascending=False)
             .head(12))  # 12 to fit small plot

top_injuries = (df_plot.groupby("Injury", as_index=False)["injury_cost_eur"]
                .sum()
                .sort_values("injury_cost_eur", ascending=False)
                .head(12))

season_trend = (df_plot.groupby("Season", as_index=False)
                .size()
                .rename(columns={"size": "injury_count"}))

season_order = sorted(season_trend["Season"].unique(),
                      key=lambda s: int(str(s).split("/")[0]))

season_trend["Season"] = pd.Categorical(season_trend["Season"], categories=season_order, ordered=True)
season_trend = season_trend.sort_values("Season")

season_league_trend = (df_plot.groupby(["Season", "league"], as_index=False)
                       .size()
                       .rename(columns={"size": "injury_count"}))
season_league_trend["Season"] = pd.Categorical(season_league_trend["Season"], categories=season_order, ordered=True)
season_league_trend = season_league_trend.sort_values(["Season", "league"])

# ---- Figure layout (3x3 grid; last panel empty) ----
fig, axes = plt.subplots(3, 3, figsize=(20, 12))
fig.suptitle("Injuries & Economic Impact Dashboard", fontsize=16, y=1.02)

# Helper: small labels
def small(ax):
    ax.tick_params(axis="both", labelsize=9)
    ax.title.set_fontsize(11)
    ax.xaxis.label.set_size(9)
    ax.yaxis.label.set_size(9)

# (1) Histogram of injury cost (log-x)
ax = axes[0, 0]
sns.histplot(df_plot["injury_cost_eur"], bins=40, ax=ax)
ax.set_xscale("log")
ax.set_title("Injury Cost Distribution (log-x)")
ax.set_xlabel("Cost (EUR)")
ax.set_ylabel("Count")
small(ax)

# (2) Total cost by league
ax = axes[0, 1]
sns.barplot(data=league_cost, x="league", y="injury_cost_eur", ax=ax)
ax.set_title("Total Injury Cost by League")
ax.set_xlabel("League")
ax.set_ylabel("Total cost (EUR)")
ax.tick_params(axis="x", rotation=25)
small(ax)

# (3) Top clubs by cost (horizontal)
ax = axes[0, 2]
sns.barplot(data=top_clubs.sort_values("injury_cost_eur"), y="club", x="injury_cost_eur", ax=ax)
ax.set_title("Top Clubs by Total Cost")
ax.set_xlabel("Total cost (EUR)")
ax.set_ylabel("")
small(ax)

# (4) Top injuries by cost (horizontal)
ax = axes[1, 0]
sns.barplot(data=top_injuries.sort_values("injury_cost_eur"), y="Injury", x="injury_cost_eur", ax=ax)
ax.set_title("Top Injuries by Total Cost")
ax.set_xlabel("Total cost (EUR)")
ax.set_ylabel("")
small(ax)

# (5) Scatter: Days vs Cost (log-y)
ax = axes[1, 1]
sns.scatterplot(data=df_plot, x="Days", y="injury_cost_eur", hue="league", alpha=0.25, legend=False, ax=ax)
ax.set_yscale("log")
ax.set_title("Days vs Injury Cost (log-y)")
ax.set_xlabel("Days injured")
ax.set_ylabel("Cost (EUR)")
small(ax)

# (6) Boxplot: cost by league (log-y)
ax = axes[1, 2]
sns.boxplot(data=df_plot, x="league", y="injury_cost_eur", showfliers=False, ax=ax)
ax.set_yscale("log")
ax.set_title("Cost by League (boxplot, log-y)")
ax.set_xlabel("League")
ax.set_ylabel("Cost (EUR)")
ax.tick_params(axis="x", rotation=25)
small(ax)

# (7) Trend: injuries by season
ax = axes[2, 0]
sns.lineplot(data=season_trend, x="Season", y="injury_count", marker="o", linewidth=2, ax=ax)
ax.set_title("Injuries Trend by Season")
ax.set_xlabel("Season")
ax.set_ylabel("Count")
ax.grid(True, alpha=0.25)
ax.tick_params(axis="x", rotation=25)
small(ax)

# (8) Trend: injuries by season & league (legend outside)
ax = axes[2, 1]
line_ax = sns.lineplot(data=season_league_trend, x="Season", y="injury_count",
                       hue="league", marker="o", linewidth=2, ax=ax)
ax.set_title("Trend by Season & League")
ax.set_xlabel("Season")
ax.set_ylabel("Count")
ax.grid(True, alpha=0.25)
ax.tick_params(axis="x", rotation=25)
small(ax)

# Grab legend handles/labels from this axis, then move legend to the right of the whole figure
handles, labels = ax.get_legend_handles_labels()
ax.legend_.remove()  # remove local legend

fig.legend(handles, labels, title="League", loc="center left", bbox_to_anchor=(1.01, 0.5))

# (9) Empty panel (optional notes)
axes[2, 2].axis("off")
axes[2, 2].text(0.0, 0.9,
                f"Rows used: {len(df_plot):,}\n"
                f"Total injury cost: ‚Ç¨{df_plot['injury_cost_eur'].sum():,.0f}\n"
                f"Total salaries: ‚Ç¨{df_plot['salary_eur'].sum():,.0f}",
                fontsize=11)

plt.tight_layout(rect=[0, 0, 0.85, 0.98])  # leave space on right for legend
plt.show()
```

## Presentation of results

First, the data sources were consolidated and cleaned to ensure consistency between injuries and salaries. The descriptive analysis was based on the filtered `df_lesiones_filtrada` and then on the physical `df_lesiones_fisicas`, where records associated with illness (e.g., virus/covid/infection) were excluded to maintain a focus on comparable physical injuries. Severity was operationalized using the variable `Days`, normalized to a numeric format (`days_num`), which allowed for quantifying days of absence without relying on heterogeneous text formats.

Subsequently, to estimate the direct economic impact, salaries and injuries were integrated using a composite key (`join_key`) constructed with league, club, season, and player. Prior to the merge, systematic name normalization (removal of accents, lowercase letters, punctuation cleaning, and space collapse) and club-specific mapping (`club_map`) were applied to resolve residual discrepancies between sources. The merge was performed as a LEFT JOIN from injuries to salaries, preserving all injury records and adding salary only when a match existed. Matching coverage was explicitly documented using the match rate (associated salaries / total injuries), allowing for methodologically transparent interpretation of the economic results.

Using the final economic base (df_lesiones_final), two metrics were derived for analysis: (i) estimated daily salary (salary_daily_eur = salary_eur / 365) and (ii) estimated cost per injury (injury_cost_eur = Days √ó salary_daily_eur). These transformations allowed for the simultaneous analysis of frequency (injury record count) and severity (days off work), as well as approximating the salary component of the economic loss.

For visualization purposes, records were aggregated by season and league to construct a temporal trend of injuries (counts). To ensure accurate interpretation, a chronological order of seasons was imposed, preventing the visualization from being affected by alphabetical sorting. The trend figure by league allows for the comparison of patterns between competitions and the detection of changes in magnitude between seasons, highlighting that the number of injury records varies by league and by season, suggesting differences in competitive workload, player profiles, or contextual conditions (without attributing direct causality).

Finally, using df_plot (a refined subset for visualization), the numerical columns were validated to ensure they were indeed numerical, and missing or invalid values ‚Äã‚Äã(residual zeros) were filtered out. This guarantees that the aggregations and sums (e.g., estimated total cost) reflect correct and comparable procedures.

## Conclusion

This work demonstrates that, even with heterogeneous public data, it is possible to build a reproducible pipeline to describe injuries by league and season, quantify severity using days off, and approximate a direct economic impact based on salaries. The cleaning and standardization process was crucial: text normalization, harmonization of leagues and clubs, and the construction of a composite key allowed for the integration of sources that were not originally designed to "talk to each other." A key lesson is that, in applied analytics, much of the value lies in making data comparable and traceable, rather than in applying "complicated" techniques without quality control.

However, the approach has significant limitations. First, the economic estimate relies on incomplete salary coverage: not all injury records are associated with a salary, so the cost results are derived from a subset and should not be interpreted as the total true cost of all injuries. Second, the estimated cost (Days √ó daily_salary) primarily captures the salary component during absence, excluding bonuses, substitutions, loss of athletic performance, effects on results, commercial revenue, or medical costs; therefore, it should be understood as a conservative approximation of the economic impact. Third, potential matching errors persist: even with normalization and mapping dictionaries, differences in names (e.g., transfers, abbreviations, or club changes) may reduce the match rate or introduce imperfect matches. Finally, the analysis presented is descriptive: it identifies patterns and differences between leagues/seasons but does not establish causality.

As future improvements, the following are suggested: (i) expand and validate salary coverage using additional sources or transparent imputation strategies when salaries are missing; (ii) construct a more robust identifier per player (ideally a stable ID) to reduce name ambiguities; (iii) Incorporate additional exposure metrics (e.g., minutes played or schedule density) to normalize by workload and allow for fairer comparisons; and (iv) extend the economic analysis with scenarios that include indirect costs (e.g., cost per point lost or replacement cost), always documenting assumptions and sensitivity. Taken together, these improvements would strengthen both the accuracy of the estimate and the practical utility of the findings for clubs, analysts, and decision-makers.
